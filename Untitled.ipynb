```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
N = 1_000_000
PL = 5000
x = np.random.random(N) * 2 - 1
y = np.random.random(N) * 2 - 1
d = x ** 2 + y ** 2
d **= 0.5
col = np.where(d > 1, 'r', 'b')
plt.figure(figsize=(9, 9))
plt.scatter(x[::N//PL], y[::N//PL], color=col[::N//PL], linewidth=0.01)
plt.show()
(d < 1).sum() * 4 / N
```

```python
c = 0
s = 0
while True:
    c += 1
    if not c % N:
        print(s / c * 4)
    x, y = np.random.random(2) * 2 - 1
    s += (x ** 2 + y ** 2) ** 0.5 <= 1
```
